```python
#Cleaning Column Names
df.columns = [c.strip() for c in df.columns]
```

```python
#Replacing the blank values with NaN
df.replace(['',' ','NA','N/A','NaN','nan','None','none'],np.nan,inplace=True)
```

```python
#Filling the missing values in text columns with Not Available
text_cols = df.select_dtypes(include='object').columns
for col in text_cols:
    df[col].fillna('Not Available',inplace=True)

#Clean 'Duration' column
if 'Duration' in df.columns:
    df['Duration'] = df['Duration'].astype(str).str.strip()
    df['duration_int'] = df['Duration'].str.extract(r'(\d+)').astype(float)
    df['duration_unit'] = df['Duration'].apply(
        lambda x: 'Season' if 'Season' in str(x) else 'Minute' if 'min' in str(x) else 'Other'
    )

#  Parse 'Release_Date' column
if 'Release_Date' in df.columns:
    df['Release_Date'] = pd.to_datetime(df['Release_Date'], errors='coerce')
    df['release_year'] = df['Release_Date'].dt.year

# Clean 'Country' and 'Type' columns
if 'Country' in df.columns:
    df['Country'] = df['Country'].astype(str).str.strip()
    df['Country_list'] = df['Country'].apply(lambda x: [c.strip() for c in x.split(',')] if pd.notna(x) else [])

if 'Type' in df.columns:
    df['Type'] = df['Type'].astype(str).str.strip()
    df['Genres_list'] = df['Type'].apply(lambda x: [t.strip() for t in x.split(',')] if pd.notna(x) else [])
df
```

```python
#Exploratory Data Analysis.
```

```python
# Count of unique titles
print("Unique Titles:", df['Title'].nunique() if 'Title' in df.columns else "N/A")

# Count of Movies vs TV Shows
if 'Category' in df.columns:
    print("\nMovies vs TV Shows:\n", df['Category'].value_counts())

# Count of Ratings
if 'Rating' in df.columns:
    print("\nRatings Distribution:\n", df['Rating'].value_counts())

# Most common countries
if 'Country' in df.columns:
    print("\nTop 10 Countries:\n", df['Country'].value_counts().head(10))

# Top Genres
if 'Type' in df.columns:
    genres = df['Type'].dropna().apply(lambda x: x.split(','))
    flat_genres = [g.strip() for sublist in genres for g in sublist]
    genre_counts = pd.Series(Counter(flat_genres)).sort_values(ascending=False)
    print("\nTop 10 Genres:\n", genre_counts.head(10))
```

```python

# Visual Analysis.

sns.set(style="whitegrid", palette="Set2")

# Content Type Distribution 
if 'Category' in df.columns:
    plt.figure(figsize=(6,4))
    sns.countplot(data=df, x='Category')
    plt.title("Distribution of Movies vs TV Shows")
    plt.xlabel("Category")
    plt.ylabel("Count")
    plt.show()
```

```python
#  Ratings Distribution 
if 'Rating' in df.columns:
    plt.figure(figsize=(8,6))
    sns.countplot(data=df, y='Rating', order=df['Rating'].value_counts().index)
    plt.title("Distribution of Content Ratings")
    plt.xlabel("Count")
    plt.ylabel("Rating")
    plt.show()
```

```python
# Top Countries
if 'Country' in df.columns:
    plt.figure(figsize=(10,6))
    sns.barplot(x=df['Country'].value_counts().head(10).values,
                y=df['Country'].value_counts().head(10).index)
    plt.title("Top 10 Countries with Most Content")
    plt.xlabel("Number of Titles")
    plt.ylabel("Country")
    plt.show()
```

```python
#Top Genres
if 'Type' in df.columns:
    plt.figure(figsize=(10,6))
    sns.barplot(x=genre_counts.head(10).values, y=genre_counts.head(10).index)
    plt.title("Top 10 Genres on Netflix")
    plt.xlabel("Count")
    plt.ylabel("Genre")
    plt.show()
```

```python
#Duration Analysis
if 'duration_int' in df.columns:
    plt.figure(figsize=(8,5))
    sns.histplot(df['duration_int'].dropna(), bins=30, kde=False)
    plt.title("Distribution of Duration (Minutes / Seasons)")
    plt.xlabel("Duration Value")
    plt.ylabel("Count")
    plt.show()
```

```python
#Duration by Category
if 'Category' in df.columns and 'duration_int' in df.columns:
    plt.figure(figsize=(8,5))
    sns.boxplot(data=df, x='Category', y='duration_int')
    plt.title("Duration by Category (Movies vs TV Shows)")
    plt.xlabel("Category")
    plt.ylabel("Duration")
    plt.show()
```
